## Dictionaries (Sözlükler)


*   Anahtar: değer mantığıyla çalışır, örn:"Türkiye":"Ankara"
*   Süslü parantezle tanımlanır {}



```python
# anahtar string olmak zorunda
# değer ise herhangi bir veri tipi olabilir
ogrenci={
    "ad":"nuh",
    "yas":35,
    "not":95.5
    }
```

```python
print(ogrenci["ad"]) #öğrencinin adına ulaşmak için
```

```python
# burada ögrenci anahtarının içinde kaç adet veri var ulaşmak için
print(len(ogrenci))
```

```python
# yeni değer eklemek istersek
ogrenci["bolum"]= "Yapay Zeka"
print(ogrenci)
```

```python
# güncelleme
ogrenci["not"]=99.9
print(ogrenci)
```

```python
# sözlükte veri silmek için del() komutu kullanılır
del ogrenci["ad"]
print(ogrenci)
```

```python
# Anahtarlar ,değerler ve tüm çiftleri görmek için keys() komutu kullanılır
print(ogrenci.keys())
print(ogrenci.values())# tüm değerleri görmek için values() komutu kullanılır
print(ogrenci.items())# tüm anahtar değer çiftleri görmek için items()komutu kullanılır
```

# **Operatörler**


*   Aritmetik operatörler

*   Karşılaştırma operatörleri

*   Mantıksal Operatörler



## Aritmetik Operatörleri


*   +,-,*,/,//,%




```python
x=10
y=3
print(f"Toplam :{x+y}")
print(f"Çıkarma :{x-y}")
print(f"Çarpma :{x*y}")
print(f"Bölme :{x/y}")
print(f"Tamsayı Bölme :{x//y}")
print(f"Kalan :{x%y}")
```

## Karşılaştırma operatörleri


*   ==,!=,<,>,<=,>=




```python
a=5
b=7
print(f"a==b :{a == b}") # a eşittir b mi a== b
print(f" a!=b :{a!=b}") # a eşit değildir b a!=b
print(f"a<b :{a<b}") # a b den küçük mü a<b
print(f"a>b :{a>b}") # a b den büyük mü a>b # a küçük b a< b
print(f"a<=b :{a<=b}") # a b den küçük eşit mi a<=b # a büyük eşit b midir a<=b
```

## Mantıksal operatörler

*   and
*   or
*   not



```python
x =5
y=10
print(f"and (ve):{x>0 and y>0}")# true ve true ise true döner
print(f"or (veya):{x>0 or y<0}")#tru yada false true olur
print(f"not (değil): {not(x>0)}")# x büyük değildir sıfırdan false döner
```
