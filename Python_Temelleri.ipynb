## Aritmetik Operatörleri

```python
x=10
y=3
print(f"Toplam :{x+y}")
print(f"Çıkarma :{x-y}")
print(f"Çarpma :{x*y}")
print(f"Bölme :{x/y}")
print(f"Tamsayı Bölme :{x//y}")
print(f"Kalan :{x%y}")
```

## Karşılaştırma operatörleri


*   ==,!=,<,>,<=,>=




```python
a=5
b=7
print(f"a==b :{a == b}") # a eşittir b mi a== b
print(f" a!=b :{a!=b}") # a eşit değildir b a!=b
print(f"a<b :{a<b}") # a b den küçük mü a<b
print(f"a>b :{a>b}") # a b den büyük mü a>b # a küçük b a< b
print(f"a<=b :{a<=b}") # a b den küçük eşit mi a<=b # a büyük eşit b midir a<=b
```

## Mantıksal operatörler

*   and
*   or

```python
x =5
y=10
print(f"and (ve):{x>0 and y>0}")# true ve true ise true döner
print(f"or (veya):{x>0 or y<0}")#tru yada false true olur
print(f"not (değil): {not(x>0)}")# x büyük değildir sıfırdan false döner
```

```python
def ortalama(a:float, b:float) -> float:
  """
  iki sayının aritmetik ortalamasını döndürür
  Inputs:
    a: float
    b: float
  output:
    average: ortalama değer
  """
  average=(a+b)/2
  return average
```

```python
ortalama(15,8)
```

```python
def ortalama_ve_max(a:float, b:float, ):
  ortalama=(a+b)/2
  max_number=max(a,b)
  return ortalama, max_number
```

```python
ortalama_deger,max_deger=ortalama_ve_max(23,9)
print(ortalama_deger)
print(max_deger)
```

```python
def faktoriyel(n:int) -> int:
  sonuc=1
  for i in range(2,n+1):
    sonuc *=i
  return(sonuc)
```

# Dosya işlemleri (I/O input-output )Operasyonları
* Dosya açma ve kapama
* Dosya yazma ve dosyadan okuma
* with yapısı

```python
# --- 1. KOD: DOSYA AÇMA VE YAZMA ---
# Dosyayı "w" modunda açıp yazıyoruz
dosya = open("ornek.txt", "w", encoding="utf-8")
dosya.write("merhaba Yapay Zeka\n")
dosya.write(" python ile dosya işleme")
dosya.close()  # Yazma işlemi bitti, dosyayı kaydettik ve kapattık.


```

```python
# --- 2. KOD: DOSYA OKUMA ---
# Aynı dosyayı bu sefer "r" modunda açıp okuyoruz
dosya = open("ornek.txt", "r", encoding="utf-8")
print("--- read ile tüm içerik ---")
print(dosya.read())
dosya.close()  # Okuma işlemi bitti, dosyayı kapattık.
```

```python
# satır satır okuma
dosya = open("ornek.txt", "r", encoding="utf-8")

print(dosya.readline())
print(dosya.readline())
dosya.close()
```

```python
# with
with open("ornek.txt", "r", encoding="utf-8") as dosya:
  print(dosya.read())
```
